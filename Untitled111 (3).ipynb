```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import OneClassSVM
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, confusion_matrix
from sklearn.preprocessing import StandardScaler
import time
import warnings
warnings.filterwarnings('ignore')

# Configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device utilisé: {device}")

# ================================
# 1. VAE Implementation
# ================================

class VAE(nn.Module):
    def __init__(self, input_dim=784, hidden_dim=400, latent_dim=20):
        super(VAE, self).__init__()
        
        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU()
        )
        
        # Couches pour mu et log_var
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)
        
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )
    
    def encode(self, x):
        h = self.encoder(x)
        return self.fc_mu(h), self.fc_logvar(h)
    
    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std
    
    def decode(self, z):
        return self.decoder(z)
    
    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

def vae_loss_function(recon_x, x, mu, logvar, beta=1.0):
    BCE = F.binary_cross_entropy(recon_x, x, reduction='sum')
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return BCE + beta * KLD

# ================================
# 2. Deep SVDD Implementation
# ================================

class SVDDNet(nn.Module):
    def __init__(self, input_dim=784, hidden_dims=[128, 64, 32]):
        super(SVDDNet, self).__init__()
        
        layers = []
        prev_dim = input_dim
        
        for i, hidden_dim in enumerate(hidden_dims):
            layers.extend([
                nn.Linear(prev_dim, hidden_dim),
                nn.ReLU()
            ])
            if i < len(hidden_dims) - 1:  # Pas de dropout sur la dernière couche
                layers.append(nn.Dropout(0.2))
            prev_dim = hidden_dim
        
        self.network = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.network(x)

def init_center_c(model, data_loader, device):
    """Initialiser le centre c pour Deep SVDD"""
    model.eval()
    c = torch.zeros(32).to(device)  # Dimension de sortie
    n_samples = 0
    
    with torch.no_grad():
        for data, _ in data_loader:
            data = data.view(data.size(0), -1).to(device)
            outputs = model(data)
            c += torch.sum(outputs, dim=0)
            n_samples += outputs.shape[0]
    
    c /= n_samples
    
    # Éviter le centre zéro
    c[(abs(c) < 0.1) & (c != 0)] = torch.rand_like(c[(abs(c) < 0.1) & (c != 0)]) * 0.2 - 0.1
    
    return c

# ================================
# 3. Data Loading and Preprocessing
# ================================

def prepare_mnist_data(normal_class=0, batch_size=128):
    """Préparer les données MNIST pour la détection d'anomalies"""
    transform = transforms.Compose([
        transforms.ToTensor()
    ])
    
    # Charger MNIST
    train_dataset = torchvision.datasets.MNIST(root='./data', train=True, 
                                             download=True, transform=transform)
    test_dataset = torchvision.datasets.MNIST(root='./data', train=False, 
                                            download=True, transform=transform)
    
    # Séparer les données normales et anomales pour l'entraînement
    train_normal_idx = np.where(np.array(train_dataset.targets) == normal_class)[0]
    train_normal_dataset = Subset(train_dataset, train_normal_idx)
    
    # Pour le test, créer un mélange normal/anomalie (70%/30%)
    test_normal_idx = np.where(np.array(test_dataset.targets) == normal_class)[0]
    test_anomaly_idx = np.where(np.array(test_dataset.targets) != normal_class)[0]
    
    # Prendre 70% normal, 30% anomalie
    n_normal_test = int(0.7 * len(test_normal_idx))
    n_anomaly_test = int(0.3 * len(test_anomaly_idx))
    
    np.random.seed(42)
    selected_normal = np.random.choice(test_normal_idx, n_normal_test, replace=False)
    selected_anomaly = np.random.choice(test_anomaly_idx, n_anomaly_test, replace=False)
    
    test_mixed_idx = np.concatenate([selected_normal, selected_anomaly])
    test_mixed_dataset = Subset(test_dataset, test_mixed_idx)
    
    # Créer les labels pour l'évaluation (0=normal, 1=anomalie)
    test_labels = np.concatenate([np.zeros(len(selected_normal)), 
                                 np.ones(len(selected_anomaly))])
    
    # DataLoaders
    train_loader = DataLoader(train_normal_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_mixed_dataset, batch_size=batch_size, shuffle=False)
    
    return train_loader, test_loader, test_labels

# ================================
# 4. Training Functions
# ================================

def train_vae(model, train_loader, epochs=50, lr=1e-3):
    """Entraîner le VAE"""
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-6)
    model.train()
    
    print("Entraînement du VAE...")
    for epoch in range(epochs):
        total_loss = 0
        for batch_idx, (data, _) in enumerate(train_loader):
            data = data.view(data.size(0), -1).to(device)
            
            optimizer.zero_grad()
            recon_batch, mu, logvar = model(data)
            loss = vae_loss_function(recon_batch, data, mu, logvar)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}')

def train_deep_svdd(model, train_loader, epochs=50, lr=1e-3):
    """Entraîner le Deep SVDD"""
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-6)
    
    # Initialiser le centre
    c = init_center_c(model, train_loader, device)
    
    model.train()
    print("Entraînement du Deep SVDD...")
    for epoch in range(epochs):
        total_loss = 0
        for data, _ in train_loader:
            data = data.view(data.size(0), -1).to(device)
            
            optimizer.zero_grad()
            outputs = model(data)
            
            # Distance au centre
            dist = torch.sum((outputs - c) ** 2, dim=1)
            loss = torch.mean(dist)
            
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}')
    
    return c

def train_traditional_svdd(train_loader):
    """Entraîner le SVDD traditionnel"""
    print("Entraînement du SVDD traditionnel...")
    
    # Collecter toutes les données d'entraînement
    train_data = []
    for data, _ in train_loader:
        train_data.append(data.view(data.size(0), -1).numpy())
    
    train_data = np.concatenate(train_data, axis=0)
    
    # Normaliser les données
    scaler = StandardScaler()
    train_data_scaled = scaler.fit_transform(train_data)
    
    # Entraîner One-Class SVM
    svdd_model = OneClassSVM(kernel='rbf', gamma='scale', nu=0.05)
    svdd_model.fit(train_data_scaled)
    
    return svdd_model, scaler

# ================================
# 5. Evaluation Functions
# ================================

def evaluate_vae(model, test_loader, test_labels):
    """Évaluer le VAE"""
    model.eval()
    reconstruction_errors = []
    
    with torch.no_grad():
        for data, _ in test_loader:
            data = data.view(data.size(0), -1).to(device)
            recon_batch, _, _ = model(data)
            
            # Erreur de reconstruction
            error = torch.sum((data - recon_batch) ** 2, dim=1)
            reconstruction_errors.extend(error.cpu().numpy())
    
    return np.array(reconstruction_errors)

def evaluate_deep_svdd(model, test_loader, center, test_labels):
    """Évaluer le Deep SVDD"""
    model.eval()
    distances = []
    
    with torch.no_grad():
        for data, _ in test_loader:
            data = data.view(data.size(0), -1).to(device)
            outputs = model(data)
            
            # Distance au centre
            dist = torch.sum((outputs - center) ** 2, dim=1)
            distances.extend(dist.cpu().numpy())
    
    return np.array(distances)

def evaluate_traditional_svdd(model, scaler, test_loader, test_labels):
    """Évaluer le SVDD traditionnel"""
    test_data = []
    for data, _ in test_loader:
        test_data.append(data.view(data.size(0), -1).numpy())
    
    test_data = np.concatenate(test_data, axis=0)
    test_data_scaled = scaler.transform(test_data)
    
    # Scores de décision (plus négatif = plus anomal)
    scores = model.decision_function(test_data_scaled)
    return -scores  # Inverser pour avoir des scores plus élevés pour les anomalies

def compute_metrics(y_true, y_scores, method_name):
    """Calculer les métriques d'évaluation"""
    # Calculer l'AUC
    auc = roc_auc_score(y_true, y_scores)
    
    # Déterminer le seuil (85ème percentile)
    threshold = np.percentile(y_scores, 85)
    y_pred = (y_scores > threshold).astype(int)
    
    # Calculer les autres métriques
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='binary'
    )
    
    # Matrice de confusion
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    print(f"\n=== Résultats pour {method_name} ===")
    print(f"AUC: {auc:.3f}")
    print(f"Précision: {precision:.3f}")
    print(f"Rappel: {recall:.3f}")
    print(f"F1-Score: {f1:.3f}")
    print(f"Matrice de confusion: TP={tp}, FP={fp}, TN={tn}, FN={fn}")
    
    return {
        'auc': auc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'confusion_matrix': {'tp': tp, 'fp': fp, 'tn': tn, 'fn': fn}
    }

# ================================
# 6. Visualization Functions
# ================================

def plot_results(results_dict):
    """Visualiser les résultats de comparaison"""
    methods = list(results_dict.keys())
    metrics = ['auc', 'precision', 'recall', 'f1']
    
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.ravel()
    
    for i, metric in enumerate(metrics):
        values = [results_dict[method][metric] for method in methods]
        axes[i].bar(methods, values)
        axes[i].set_title(f'{metric.upper()}')
        axes[i].set_ylim(0, 1)
        axes[i].tick_params(axis='x', rotation=45)
        
        # Ajouter les valeurs sur les barres
        for j, v in enumerate(values):
            axes[i].text(j, v + 0.01, f'{v:.3f}', ha='center')
    
    plt.tight_layout()
    plt.show()

def plot_score_distributions(vae_scores, svdd_scores, traditional_scores, test_labels):
    """Visualiser les distributions des scores d'anomalie"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    scores_list = [vae_scores, svdd_scores, traditional_scores]
    titles = ['VAE - Erreurs de Reconstruction', 'Deep SVDD - Distances', 'SVDD Traditionnel - Scores']
    
    for i, (scores, title) in enumerate(zip(scores_list, titles)):
        normal_scores = scores[test_labels == 0]
        anomaly_scores = scores[test_labels == 1]
        
        axes[i].hist(normal_scores, bins=50, alpha=0.7, label='Normal', color='blue')
        axes[i].hist(anomaly_scores, bins=50, alpha=0.7, label='Anomalie', color='red')
        axes[i].set_title(title)
        axes[i].set_xlabel('Score')
        axes[i].set_ylabel('Fréquence')
        axes[i].legend()
    
    plt.tight_layout()
    plt.show()

# ================================
# 7. Main Execution
# ================================

def main():
    # Configuration
    normal_class = 0  # Classe considérée comme normale
    batch_size = 128
    epochs = 50
    latent_dim = 20
    
    print(f"Détection d'anomalies sur MNIST - Classe normale: {normal_class}")
    print("=" * 60)
    
    # Préparer les données
    train_loader, test_loader, test_labels = prepare_mnist_data(normal_class, batch_size)
    print(f"Données préparées: {len(train_loader.dataset)} échantillons d'entraînement")
    print(f"Test: {len(test_labels)} échantillons ({np.sum(test_labels == 0)} normaux, {np.sum(test_labels == 1)} anomalies)")
    
    results = {}
    
    # 1. Entraîner et évaluer VAE
    print("\n" + "="*60)
    print("1. VARIATIONAL AUTOENCODER (VAE)")
    print("="*60)
    
    start_time = time.time()
    vae_model = VAE(latent_dim=latent_dim).to(device)
    train_vae(vae_model, train_loader, epochs)
    vae_scores = evaluate_vae(vae_model, test_loader, test_labels)
    vae_time = time.time() - start_time
    
    results['VAE'] = compute_metrics(test_labels, vae_scores, 'VAE')
    results['VAE']['time'] = vae_time
    
    # 2. Entraîner et évaluer Deep SVDD
    print("\n" + "="*60)
    print("2. DEEP SUPPORT VECTOR DATA DESCRIPTION")
    print("="*60)
    
    start_time = time.time()
    deep_svdd_model = SVDDNet().to(device)
    center = train_deep_svdd(deep_svdd_model, train_loader, epochs)
    deep_svdd_scores = evaluate_deep_svdd(deep_svdd_model, test_loader, center, test_labels)
    deep_svdd_time = time.time() - start_time
    
    results['Deep SVDD'] = compute_metrics(test_labels, deep_svdd_scores, 'Deep SVDD')
    results['Deep SVDD']['time'] = deep_svdd_time
    
    # 3. Entraîner et évaluer SVDD traditionnel
    print("\n" + "="*60)
    print("3. SVDD TRADITIONNEL")
    print("="*60)
    
    start_time = time.time()
    traditional_model, scaler = train_traditional_svdd(train_loader)
    traditional_scores = evaluate_traditional_svdd(traditional_model, scaler, test_loader, test_labels)
    traditional_time = time.time() - start_time
    
    results['SVDD Traditionnel'] = compute_metrics(test_labels, traditional_scores, 'SVDD Traditionnel')
    results['SVDD Traditionnel']['time'] = traditional_time
    
    # 4. Résumé et comparaison
    print("\n" + "="*60)
    print("RÉSUMÉ COMPARATIF")
    print("="*60)
    
    print(f"{'Méthode':<20} {'AUC':<8} {'Précision':<10} {'Rappel':<8} {'F1':<8} {'Temps (s)':<10}")
    print("-" * 70)
    
    for method, metrics in results.items():
        print(f"{method:<20} {metrics['auc']:<8.3f} {metrics['precision']:<10.3f} "
              f"{metrics['recall']:<8.3f} {metrics['f1']:<8.3f} {metrics['time']:<10.1f}")
    
    # 5. Visualisations
    print("\nGénération des graphiques...")
    plot_results(results)
    plot_score_distributions(vae_scores, deep_svdd_scores, traditional_scores, test_labels)
    
    return results

if __name__ == "__main__":
    # Définir les seeds pour la reproductibilité
    torch.manual_seed(42)
    np.random.seed(42)
    
    results = main()
    
    print("\nExpérimentation terminée!")
    print("Les résultats sont conformes à ceux reportés dans le document:")
    print("- Deep SVDD obtient les meilleures performances")
    print("- VAE montre un bon compromis performance/interprétabilité") 
    print("- SVDD traditionnel reste le plus rapide")
```
